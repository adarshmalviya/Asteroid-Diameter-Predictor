## Import Necessary Liberaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns', 50) # To see all the columns of dataset
```

## Import DataSet

```python
dataset = pd.read_csv('Clean_Dataset.csv')  # To Read dataset
dataset.head()
```

```python
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset.head()
```

```python
# Segregate Data into numeric and categorical onces
categorical, numeric = [], []
for ele in dataset.columns:
    if dataset[ele].dtype == 'object':
        categorical.append(ele)
    else:
        numeric.append(ele)
print("Categorical Data : ", categorical)
print("Numeric Data : ", numeric)
```

## Data Visualization

```python
plt.figure(figsize=(12,12))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

### Feature 'n' is having negative correlation with many other features

```python
plt.scatter(dataset['a'], dataset['n'])
plt.title("a VS n")
plt.xlabel("a values")
plt.ylabel("n values")
plt.show()
```

### A perfect Negative Correlation is visible

```python
plt.scatter(dataset['n'], dataset['per_y'])
plt.title("n VS per_y")
plt.xlabel("n values")
plt.ylabel("per_y values")
plt.show()
```

### Box plot

### kde Plot

### Histogram
